### Упражение 2. Вычисление нормы (модуля) вектора

```python
# Создадим функцию, вычисляющую модуль вектора
def NormOfVector(v):
    return np.sqrt(sum(v**2))

# Проверим на единичном векторе
print(NormOfVector(np.array([1, 0, 0, 0])))

# Возьмем не единичный вектор и сравним с фуккцией np.linalg.norm()
w = np.array([1, 2, 3])
print(NormOfVector(w), np.linalg.norm(w))
```

### Упражение 3. Функция, делающая из вектора единичный вектор в том же направлении

```python
# Создадим данную функцию
def createUnitVector(v):
    mu = np.linalg.norm(v) # Получим модуль вектора
    return v / mu # возращаем единичный вектор

# Протестируем функцию на единичном векторе
w = np.array([0, 0, 1])
print(createUnitVector(w))

# Протестируем функцию на неединичном, но "легком векторе"
v = np.array([0, 0, 5])
print(createUnitVector(v))

# Протестируем на не единичном векторе
a = np.array([23, -12, 5])
ua = createUnitVector(a)
print(ua), print("  ")
# Подтвердим нормы векторов
print(np.linalg.norm(a), np.linalg.norm(ua))

# Что произойдет с нуль-вектором?
createUnitVector(np.zeros((1,4)))
```

### Упражнение 4. Функция, делющая из вектора другой вектор, в том же направлении, но определенной нормы

```python
# Создадим требуемую функцию
def createMagVector(v, mag):
    mu = np.linalg.norm(v) # вычислим норму вектора
    return mag * v / mu # Умножаем единичный вектор на заданную длину, получая масштабированный ветор

# проверим вектор, который легко представить
w = np.array([0, 0, 1])
mw = createMagVector(w, 4)
print(mw)
    
# Подтвердим нормы векторов
print(np.linalg.norm(w), np.linalg.norm(mw))
```

### Упражнение 7. Код, демонстрирующий, что скалярное произведение коммутативно. ($a^{Т}b=b^{Т}a$)

```python
# Зададим размерность
n = 11

# Создадим два вектора-столбца
a = np.random.randn(n, 1)
b = np.random.randn(n, 1)

# Запишем скалярное произведение в "обоих направлениях"
atb = sum(a*b)
bta = sum(b*a)

# Если их разность равна нулю, то они равны:
print(atb - bta)
```

```python
# В качестве дополнительной задачи посмотрим, что происходит когда мы используем dot
adb = np.dot(a.T, b)
bda = np.dot(b.T, a)

print(adb - bda)

# Однако, если мы переставим столбец и строку местами, то у нас получится внешнее произведение,
# которое в общем случае не коммутативно
adb_2 = np.dot(a, b.T)
bda_2 = np.dot(b, a.T)
print(adb_2, "\n\n", bda_2, "\n\n", adb_2-bda_2)
```

### Упражение 9. Реализовать ортагональное разложение векторов.

```python
# Задаем векторы
r = np.random.randn(2) # Опорный вектор
t = np.random.randn(2) # Целевой вектор

# Вычислим паралельную компоненту вектора t
beta = np.dot(r, t) / np.dot(r, r)
tPara = beta * r

#Вычислим перпендикуляреую компоненту
tPerp = t - tPara

# Проверим, что компоненты в сумме равны целевому вектору
print(t)
print(tPara+tPerp)

# Проверим, что скалярное (точечное) произведение компонент равна нулю:
print(np.dot(tPara, tPerp))

# Изобразим разложение
plt.figure(figsize=(6,6))

# Добавим главные вектора
plt.plot([0, r[0]], [0, r[1]], color=[.7,.7,.7], label="$r$", linewidth=3)
plt.plot([0, t[0]], [0, t[1]], color="k", label="$t$", linewidth=3)

# Добавим компоненты на график
plt.plot([0, tPara[0]], [0, tPara[1]], "--", color="k", linewidth=3, label="$t_{\parallel}$")
plt.plot([0, tPerp[0]], [0, tPerp[1]], ":", color="k", linewidth=3, label="$t_{\perp}$")

# Наведем красоту
plt.axis("square")
plt.legend()
plt.show()
```

### Упражнение 10. Что будет, если при нахождении параллельной компоненты, вместо произведения $r^{Т}r$ вычислить $t^{Т}t$ в константе $\beta$

```python
# Задаем векторы
r = np.random.randn(2) # Опорный вектор
t = np.random.randn(2) # Целевой вектор

# Вычислим паралельную компоненту вектора t
beta = np.dot(r, t) / np.dot(t, t)
tPara = beta * r

#Вычислим перпендикуляреую компоненту
tPerp = t - tPara

# Проверим, что компоненты в сумме равны целевому вектору
print(t)
print(tPara+tPerp)

# Проверим, что скалярное (точечное) произведение компонент равна нулю:
print(np.dot(tPara, tPerp))

# Изобразим разложение
plt.figure(figsize=(6,6))

# Добавим главные вектора
plt.plot([0, r[0]], [0, r[1]], color=[.7,.7,.7], label="$r$", linewidth=3)
plt.plot([0, t[0]], [0, t[1]], color="k", label="$t$", linewidth=3)

# Добавим компоненты на график
plt.plot([0, tPara[0]], [0, tPara[1]], "--", color="k", linewidth=3, label="$t_{\parallel}$")
plt.plot([0, tPerp[0]], [0, tPerp[1]], ":", color="k", linewidth=3, label="$t_{\perp}$")

# Наведем красоту
plt.axis("square")
plt.legend()
plt.show()
```
